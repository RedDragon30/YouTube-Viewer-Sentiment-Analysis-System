# Create Your Own Dataset Using Youtube API (Optional)

```python
# Importing necessary libraries
import pandas as pd
from googleapiclient.discovery import build
```

```python
# YouTube API setup
api_key = 'YOUR_API_KEY_HERE' # Replace with your actual API key
youtube = build('youtube', 'v3', developerKey=api_key)
```

```python
# Searching for videos with the keyword like 'technology' or any other topic
request = youtube.search().list(
    part='id',
    q='technology',
    type='video',
    maxResults=100
)
response = request.execute()

video_ids = []

for item in response['items']:
    if item['id']['kind'] == 'youtube#video':
        video_ids.append(item['id']['videoId'])

video_ids
```

```python
# Fetching comments for each video
comment_list = []

for vedio_id in video_ids:
    # Comments
    comments = youtube.commentThreads().list(
        part='snippet,replies',
        videoId=vedio_id,
        maxResults=100
    ).execute()
    
    # Extracting comments into a list
    for item in comments['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        comment_list.append(comment['textDisplay'])
        print(f"Comment by {comment['authorDisplayName']}: {comment['textDisplay']}")
```

```python
# Creating a DataFrame
df = pd.DataFrame(comment_list, columns=['comment'])
df
```

```python
# Saving comments to a CSV file
df.to_csv('youtube_comments.csv', index=False)
```
